```java
import com.azure.ai.openai.OpenAIAsyncClient;
import com.azure.ai.openai.OpenAIClientBuilder;
import com.azure.ai.openai.models.Choice;
import com.azure.ai.openai.models.Completions;
import com.azure.ai.openai.models.CompletionsOptions;
import com.azure.ai.openai.models.CompletionsUsage;
```

```java
String AZURE_KEY = "Your Azure OpenAI Service API Key";
String AZURE_ENDPOINT = "Your Azure OpenAI Service Endpoint";
```

```java

import com.azure.core.credential.AzureKeyCredential;
```

```java
OpenAIAsyncClient client = new OpenAIClientBuilder()
        .credential(new AzureKeyCredential(AZURE_KEY))
        .endpoint(AZURE_ENDPOINT)
        .buildAsyncClient();
```

```java
List<String> prompt = new ArrayList<>();
prompt.add("How to learn AI");
```

```java
var completionsOptions = new CompletionsOptions(prompt);
```

```java
completionsOptions.setTemperature(0.8);
completionsOptions.setMaxTokens(10000);
```

```java
var completions = client.getCompletions("Your Azure OpenAI Service gpt-35-indtruct Deployment Name", completionsOptions);
```

```java
for (Choice choice : completions.block().getChoices()) {
    System.out.println(choice.getText());
}
```

```java
import com.azure.ai.openai.models.ChatChoice;
import com.azure.ai.openai.models.ChatCompletions;
import com.azure.ai.openai.models.ChatCompletionsOptions;
import com.azure.ai.openai.models.ChatRequestMessage;
import com.azure.ai.openai.models.ChatRequestSystemMessage;
import com.azure.ai.openai.models.ChatRequestUserMessage;
import com.azure.ai.openai.models.ChatResponseMessage;
import com.azure.ai.openai.models.ChatRole;
```

```java
List<ChatRequestMessage> chatMessages = new ArrayList<>();
chatMessages.add(new ChatRequestSystemMessage("You are a AI teacher."));
chatMessages.add(new ChatRequestUserMessage("Can you explain what is Machine Learning?"));
```

```java
var chatCompletions = client.getChatCompletions("Your Azure OpenAI Service gpt-35-turbo Deployment Name",
    new ChatCompletionsOptions(chatMessages).setMaxTokens(10000).setTemperature(0.7));
```

```java
for (ChatChoice choice : chatCompletions.block().getChoices()) {
    ChatResponseMessage message = choice.getMessage();
    System.out.println(message.getContent());
}
```
